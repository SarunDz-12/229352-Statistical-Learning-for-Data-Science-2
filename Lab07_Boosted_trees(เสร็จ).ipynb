# AdaBoots

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier

X_train = pd.read_csv("X_train.csv")
y_train = pd.read_csv("y_train.csv").values.ravel()

model = AdaBoostClassifier(random_state=42)

param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.1, 1.0]
}

gs = GridSearchCV(model, param_grid, cv=5, scoring="accuracy")
gs.fit(X_train, y_train)

print(gs.best_score_)
```

# XGBoots

```python
xg = XGBClassifier()
params = {'n_estimators':[25,50,100,150,200], 'max_depth':[1,2,3,4,5,6], 'learning_rate':[1,2,3,4]}
gridxg= GridSearchCV(xg, params, scoring='accuracy')
```

```python
gridxg.best_params_
```

```python
gridxg.best_score_
```

```python
import pandas as pd

X_test = pd.read_csv("/content/X_test.csv")
y_test = pd.read_csv("/content/y_test.csv").squeeze()

X_test = X_test.reindex(columns=X_train.columns)

best_xg = gridxg.best_estimator_

y_pred_xg = best_xg.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_xg))
```

# LGBM

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# à¸ªà¸£à¹‰à¸²à¸‡à¹‚à¸¡à¹€à¸”à¸¥
lgbm = LGBMClassifier(random_state=42)

# à¸à¸³à¸«à¸™à¸”à¸žà¸²à¸£à¸²à¸¡à¸´à¹€à¸•à¸­à¸£à¹Œ
param_grid = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.05, 0.1, 0.2],
    "max_depth": [-1, 5, 10],
    "num_leaves": [31, 50, 100]
}

# GridSearchCV
grid_lgbm = GridSearchCV(
    estimator=lgbm,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

# train
grid_lgbm.fit(X_train, y_train)
```

```python
grid_lgbm.best_score_
```

```python
best_lgbm = grid_lgbm.best_estimator_
```

```python
X_test = X_test.reindex(columns=X_train.columns)

y_pred = best_lgbm.predict(X_test)
test_acc = accuracy_score(y_test, y_pred)

print("Test accuracy:", test_acc)
```

```python
import matplotlib.pyplot as plt
from lightgbm import plot_importance

plt.figure()
plot_importance(best_lgbm, max_num_features=10)
plt.title("Feature Importances (LGBM)")
plt.show()
```
